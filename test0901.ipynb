numpy를 사용한 버블 차트 그리기

```python
import matplotlib.pyplot as plt
import numpy as np

#10~100 사이 200개 
x = np.random.randint(10, 100, 200)
y = np.random.randint(10, 100, 200)


#0~100까지 100개
size = np.random.randint(10,100, 200)

plt.scatter(x,y,s=size,c=x,cmap='jet',alpha=0.7)
plt.colorbar()
plt.show()
```

마스크 (mask) 기능 알아보기

```python
import numpy as np
a = np.arange(-5,5)
print(a)

print(a<0)

print(a[a<0])
print()
mask1 = abs(a) > 3


mask2 = abs(a) % 2 == 0

print(mask1)
print(mask2)
print('a[mask1]: ' ,a[mask1])
print('a[mask1+mask2]:', a[mask1+mask2]) #둘 중 하나의 조건이라도 참일 경우
print('a[mask1*mask2]:', a[mask1*mask2]) #두 가지 조건이 모두 참일 경우
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(-100, 100, 1000)
y = np.random.randint(-100, 100, 1000)
mask1 = abs(x) > 50 #절대값이 50보다 큰 값
mask2 = abs(y) > 50
x = x[mask1+mask2] #mask1과 mask2 중 하나라도 만족하는 값 저장
y = y[mask1+mask2]

size = np.random.random(len(x)) * 100
plt.scatter(x,y, s=size, c=x, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```
