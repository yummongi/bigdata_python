## 서울 신도림 인구 구조

```python
import csv
import matplotlib.pyplot as plt
```

```python
f = open('age.csv')

data = csv.reader(f)

result = []

for row in data:
    if '신도림' in row[0]:
        for i in row[3:]:
            result.append(int(i))

plt.style.use('ggplot')
plt.plot(result)
plt.show()
```

## 입력값에 대한 지역의 인구 구조

```python
f = open('age.csv')

data = csv.reader(f)

result = []

city = input("인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력하세요.")

for row in data:
    if city in row[0]:
        for i in row[3:]:
            result.append(int(i.replace(',','')))

plt.style.use('ggplot')
plt.rc('font', family = 'Malgun Gothic')
plt.title(city + ' 지역의 인구 구조')
plt.plot(result)
plt.show()
```
