## 서울 신도림 인구 구조 -  plot()

```python
import csv
import matplotlib.pyplot as plt
```

```python
f = open('age.csv')

data = csv.reader(f)

result = []

for row in data:
    if '신도림' in row[0]:
        for i in row[3:]:
            result.append(int(i))

plt.style.use('ggplot')
plt.plot(result)
plt.show()
```

## 입력값에 대한 지역의 인구 구조 - plot()

```python
f = open('age.csv')

data = csv.reader(f)

result = []

city = input("인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력하세요.")

for row in data:
    if city in row[0]:
        for i in row[3:]:
            result.append(int(i.replace(',','')))

plt.style.use('ggplot')
plt.rc('font', family = 'Malgun Gothic')
plt.title(city + ' 지역의 인구 구조')
plt.plot(result)
plt.show()
```

## 남, 여 지역별 인구 구조 - gender.csv

```python
f = open('gender.csv')

data = csv.reader(f)

male = []
female = []

city = input("인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력하세요.")

for row in data:
    if city in row[0]:
        for i in row[3:104]:
            male.append(-int(i))
        for i in row[106:]:
            female.append(int(i))

import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.figure(figsize=(10,5), dpi=300)
plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False

plt.title(city + ' 지역의 인구 구조')
plt.barh(range(101),male, color = 'blue', label = '남성')
plt.barh(range(101),female, color = 'red', label = '여성')
plt.legend()
plt.show()
```

```python
f = open('gender.csv')

data = csv.reader(f)

male = []
female = []

city = input("인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력하세요.")

for row in data:
    if city in row[0]:
        for i in range(0, 101):
            male.append(-int(row[(i+3)]))
            female.append(int(row[-(i+1)]))
female.reverse()

import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.figure(figsize=(10,5), dpi=300)
plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False

plt.title(city + ' 지역의 인구 구조')
plt.barh(range(101),male, color = 'blue', label = '남성')
plt.barh(range(101),female, color = 'red', label = '여성')
plt.legend()
plt.show()
```
