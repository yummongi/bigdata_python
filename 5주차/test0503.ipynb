## 대중 교통 유임 승차 비율이 가장 높은 역 

```python
import csv
import matplotlib.pyplot as plt

f = open('subwayfee.csv')
data = csv.reader(f)

#유임승차/무임승차 = 유임승차 비율 (높을 수록 유임승차 많음)
rate = 0

#최대값
mx = 0

#헤더
next(data)

for row in data:
    #4~8열까지 정수로 변환
    for i in range(4,8):
        row[i] = int(row[i])
    #무임승차에 0값이 있을 수 있다.
    if row[6] != 0:
        rate = row[4] / row[6]
        if (rate > mx):
            mx = rate
            print(row, round(mx,2))
    
```

변경 rate = 유임 승차 인원 / 전체 (유임 + 무임 ) 인원

```python
import csv
import matplotlib.pyplot as plt

f = open('subwayfee.csv')
data = csv.reader(f)
#헤더
next(data)

#유임승차/무임승차 = 유임승차 비율 (높을 수록 유임승차 많음)
rate = 0

#최대값
mx = 0

mx_station = 0

for row in data:
    #4~8열까지 정수로 변환
    for i in range(4,8):
        row[i] = int(row[i])
    #무임승차에 0값이 있을 수 있다.
    if row[6] != 0 and (row[4] + row[6]) > 100000:
        rate = row[4] / (row[4] + row[6])
        if (rate > mx):
            mx = rate
            mx_station = row[3] + ' ' + row[1]
            
print(mx_station, round(mx*100,2))
    
```
