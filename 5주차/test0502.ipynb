## 남녀 비율을 더 알아보기 쉽게 표현하기

 꺾은선 그래프로 인구 구조 표현

```python
import csv
import matplotlib.pyplot as plt
```

```python
f = open('gender.csv')
data = csv.reader(f)

m = []
f = []

name = input('궁금한 동네를 입력해주세요: ')
for row in data:
    if name in row[0]:
        for i in range(3,104):
            m.append(int(row[i].replace(',','')))
            f.append(int(row[i+103].replace(',','')))
        break
        
plt.plot(m, label = 'Male')
plt.plot(f, label = 'Female')
plt.legend()
plt.show()
```

꺾은선 그래프로 인구 구조 표현

```python
f = open('gender.csv')
data = csv.reader(f)

result = []

name = input('궁금한 동네를 입력해주세요: ')
for row in data:
    if name in row[0]:
        for i in range(3,104):
            result.append(int(row[i].replace(',','')) - int(row[i+103].replace(',','')))
        break
        
plt.bar(range(101), result)
plt.show()
```

```python
#colorbar() 함수를 사용하면 그래프 옆에 컬러바를 추가할 수 있다.
#c 속성을 추가하여 표현하고 싶은 색상의 개수를 설정하면 해당하는 컬러바의 색으로 정해진다.
plt.scatter([1,2,3,4], [10,30,20,40], s=[100,200,300,400], c= range(4))
plt.colorbar()
plt.show()
```

```python
#cmap 컬러바에 사용될 색상의 종류를 정할 수 있다. 무지개색 - jet
#https://matplotlib.org/stable/tutorials/colors/colormaps.html
plt.scatter([1,2,3,4], [10,30,20,40], s=[100,200,300,400], c= range(4), cmap = 'jet')
plt.colorbar()
plt.show()
```

```python
import random

x = []
y = []
size = []

for i in range(100):
    x.append(random.randint(50,100))
    y.append(random.randint(50,100))
    size.append(random.randint(10,100))

#뒤에 작은점이 가려지기때문에 alpha로 조절한다 0~1
plt.scatter(x, y, s=size, c=size, cmap='jet', alpha=0.7)
plt.show()
```

연령별 남성과 여성 인구수를 합친 값을 size 에 넣어 점의 크기를 표현

```python
import math

f = open('gender.csv')
data = csv.reader(f)

size = []
m = []
f = []
name = input('궁금한 동네를 입력하세요: ')
for row in data:
    if name in row[0]:
        for i in range(3,104):
            m.append(int(row[i].replace(',','')))
            f.append(int(row[i+103].replace(',','')))
            size.append(math.sqrt(int(row[i].replace(',','')))+int(row[i+103].replace(',','')))
        break
#그리드 배경
plt.style.use('ggplot')
#폰트 변경
plt.rc('font', family='Malgun Gothic')
#그래프 크기 변경
plt.figure(figsize=(10,5), dpi=300)
plt.title(name + '지역의 성별 인구 그래프')
plt.scatter(m,f, s=size, c=range(101), cmap='jet', alpha=0.5)
plt.colorbar()
plt.plot(range(max(m)), range(max(m)), color='green') #추세선 춫가
plt.xlabel('남성 인구수')
plt.ylabel('여성 인구수')
plt.show()
```
