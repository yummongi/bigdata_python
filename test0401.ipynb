## 서울 최고 기온 데이터 - boxplot

```python
f = open('seoul.csv')
data = csv.reader(f)

result = []

next(data)

for row in data:
    if row[-1] != '':
        result.append(float(row[-1]))

plt.boxplot(result)
plt.show()
```

## 1월과 8월 서울 최고 기온 - boxplot

```python
f = open('seoul.csv')
data = csv.reader(f)

result = []
one = []
eight = []

next(data)

for row in data:
    if row[-1] != '':
        month = row[0].split('-')[1]
        if month == '01':
            one.append(float(row[-1]))
        if month == '08':
            eight.append(float(row[-1]))
            

plt.boxplot(one)
plt.boxplot(eight)
plt.show()
```

## 서울 월별 최고 기온 데이터 - boxplot

```python
f = open('seoul.csv')
data = csv.reader(f)
next(data)

month = []
for i in range(12):
    month.append([])


for row in data:
    if row[-1] != '':
        month[int(row[0].split('-')[1])-1].append(float(row[-1]))
            

plt.boxplot(month)
plt.show()
```

## 서울 8월 일별 기온 데이터 - boxplot

```python
f = open('seoul.csv')
data = csv.reader(f)
next(data)

day = []
for i in range(31):
    day.append([])


for row in data:
    if row[-1] != '':
        date = row[0].split('-')
        if date[1] == '08':
            day[int(date[2])-1].append(float(row[-1]))

plt.style.use('ggplot')
plt.figure(figsize=(10,5), dpi= 300)
plt.boxplot(day, showfliers=False)
plt.show()
```
